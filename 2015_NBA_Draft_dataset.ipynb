##**Creating a DataFrame**


Ensure that the file "2015_nba_draft_combine.csv" is in your directory underneath sample_data.

It is important that the file name is **exactly** the same in the code as it is in the title or else the cell block will not function properly.

The command: pd.read_csv reads the data in the dataframe when set equal to df.

```python
df=pd.read_csv("2015_nba_draft_combine.csv")
```

```python
Player = ["Aaron Harrison", "Aaron White","Alan Williams","Andrew Harrison","Anthony Brown"]
```

```python
combined=pd.DataFrame({"Player":[], "HeightWithShoes":[],"Weight":[]})
```

```python
for x in Player:
  current = pd.DataFrame({"Player": [x],"HeightWithShoes":[df[df["Player"]==x].HeightWithShoes.sum()], "Weight":[df[df['Player']==x].Weight.sum()]})
  combined = pd.concat([combined,current],axis=0,ignore_index=True,sort=False)
```

```python
combined = combined.rename(columns={'HeightWithShoes':'Height With Shoes'})
combined.to_csv("NBADraftPicks-2015.csv",index=False)
```
